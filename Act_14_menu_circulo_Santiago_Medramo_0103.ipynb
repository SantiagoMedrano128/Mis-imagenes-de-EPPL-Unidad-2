### Redaccion del problema
Desarrollar un programa interactivo en Python con formulas a utilizar para calcular tres magnitudes geometricas basicas a partir del radio ingresado por el usuario: el area de un circulo, el perimetro (circunferencia) de un circulo y el volumen de una esfera. El programa debe mostrar un menu numerico que se repite hasta que el usuario decida salir, solicitando el radio unicamente cuando se selecciona una opcion de calculo. No se requiere validacion de dotos ni uso deprogramacion orientada a objetos.
### Formulas
1. $Área: 𝐴=𝜋⋅𝑟2$
2. $Perímetro:P=2⋅π⋅r$
3. $Volumen Esfera: V=4/3⋅π⋅r3$

![texto del vinculo](https://raw.githubusercontent.com/SantiagoMedrano128/Mis-imagenes-de-EPPL-Unidad-2/refs/heads/main/el%20circulo.PNG)



```python
import math

def areacirculo():
    """Calcula el área de un círculo solicitando el radio."""
    print("\n📏 --- CÁLCULO DEL ÁREA DEL CÍRCULO ---")
    radio = float(input("Por favor, ingresa el radio del círculo: "))

    # Fórmula: Área = π * r^2
    area = math.pi * (radio ** 2)

    print(f"✨ El área del círculo con radio {radio} es: {area:.2f}")

def perimetrocirculo():
    """Calcula el perímetro (circunferencia) de un círculo solicitando el radio."""
    print("\n🔄 --- CÁLCULO DEL PERÍMETRO DEL CÍRCULO ---")
    radio = float(input("Por favor, ingresa el radio del círculo: "))

    # Fórmula: Perímetro = 2 * π * r
    perimetro = 2 * math.pi * radio

    print(f"✨ El perímetro del círculo con radio {radio} es: {perimetro:.2f}")

def volumencirculo():
    """Calcula el volumen de una esfera solicitando el radio."""
    print("\n🌍 --- CÁLCULO DEL VOLUMEN DE LA ESFERA ---")
    radio = float(input("Por favor, ingresa el radio de la esfera: "))

    # Fórmula: Volumen = (4/3) * π * r^3
    volumen = (4/3) * math.pi * (radio ** 3)

    print(f"✨ El volumen de la esfera con radio {radio} es: {volumen:.2f}")

def menu():
    """Muestra el menú principal y controla el flujo del programa."""
    opcion = 0

    while opcion != 4:
        print("\n" + "="*40)
        print("   🧮 MENÚ DE GEOMETRÍA 🧮")
        print("="*40)
        print("1️⃣  Calcular Área del Círculo")
        print("2️⃣  Calcular Perímetro del Círculo")
        print("3️⃣  Calcular Volumen de la Esfera")
        print("4️⃣  Salir del programa")
        print("="*40)

        # Se asume que el usuario ingresa un número válido según los requisitos
        opcion = int(input("👉 Elige una opción (1-4): "))

        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumencirculo()
        elif opcion == 4:
            print("\n👋 ¡Gracias por usar el programa! Hasta luego.")
            print("Programa realizado por Santiago Medrano NL = 37")
        else:
            print("\n⚠️  Opción no válida. Por favor, elige entre 1 y 4.")

# Llamada inicial para iniciar el programa
if __name__ == "__main__":
    menu()
```
